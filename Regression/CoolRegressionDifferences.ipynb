```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
from ipynb.fs.full.LassoRegression import LassoRegression
from ipynb.fs.full.RidgeRegression import RidgeRegression
from ipynb.fs.full.LinearRegression import LinearRegression
```

```python
# Generate synthetic data
X, y = make_regression(n_samples=500, n_features=1, noise=15, random_state=4)

# Split the dataset into 70% training and 30% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Print the shapes of the training and testing sets
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
plt.scatter(X_train, y_train, color='blue', marker='o', label='Data points')

# Add labels and title
plt.xlabel('Feature (X)')
plt.ylabel('Target (y)')
plt.title('Scatter Plot of Generated Data')

# Display the plot
plt.show()
```

```python
lassoReg = LassoRegression(alpha=15)
ridgeReg = RidgeRegression(alpha=15)
linearReg = LinearRegression()
```

```python
ridgeReg.fit(X_train, y_train)
lassoReg.fit(X_train, y_train)
linearReg.fit(X_train, y_train)
```

```python
predictionsLasso = lassoReg.predict(X_test)
predictionsRidge = ridgeReg.predict(X_test)
predictionsLinear = linearReg.predict(X_test)
```

```python
plt.scatter(X_test, y_test, label='Original Data')
plt.plot(X_test, predictionsLinear, color='red', label='Predicted Linear Line')
plt.plot(X_test, predictionsRidge, color='green', label='Predicted Ridge Line')
plt.plot(X_test, predictionsLasso, color='orange', label='Predicted Lasso Line')

    # Add labels and title
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression: Test and Visualization')

    # Add a legend
plt.legend()

    # Display the plot
plt.show()
```

If we look closely we can see a clear divergence between the lines of ridge, lasso and linear.
